### 1.1 Normalizing numeric data, which are non IDs

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = ['reply_count', 'quote_count', 'retweet_count', 'like_count']

scaler = MinMaxScaler()

df_result[columns_to_normalize] = scaler.fit_transform(df_result[columns_to_normalize])

df_result['impression_count'] = df_result[columns_to_normalize].mean(axis=1)

df_result.head()
```

### 1.2 Recency Score

```python
from datetime import datetime
def calculate_recency_score():
	df_result['created_at'] = pd.to_datetime(df_result['created_at'])

	most_recent_date = df_result['created_at'].max()

	time_elapsed_days = (most_recent_date - df_result['created_at']).dt.total_seconds() / (60*60*24)

	lambda_value = 0.03 

	df_result['recency_score'] = np.exp(-lambda_value * time_elapsed_days)
 
calculate_recency_score()

df_result.to_csv('./Database/tweets_processed.csv', index=False)
```

### 1.3 Engagement Rate

```python
alpha, beta, gamma, delta = 0.9, 0.9, 0.9, 0.9  # Replies, Quotes, Retweets, Likes
epsilon = 0.9  # Impressions

grouped_data = df_result.groupby('author_id').agg(
    total_likes=('like_count', 'sum'),
    total_retweets=('retweet_count', 'sum'),
    total_replies=('reply_count', 'sum'),
    total_quotes=('quote_count', 'sum'), 
    total_impressions=('impression_count', 'sum'),
    recency=('recency_score', 'mean') 
).reset_index()

grouped_data['engagement_rate'] = np.sqrt(
    alpha * grouped_data['total_replies'] + 
    beta * grouped_data['total_quotes'] + 
    gamma * grouped_data['total_retweets'] + 
    delta * grouped_data['total_likes'] + 
    epsilon * np.log1p(grouped_data['total_impressions'])
) * grouped_data['recency']

max_transformed = grouped_data['engagement_rate'].max()
df_result['engagement_rate'] = grouped_data['engagement_rate'] / max_transformed
```

### Calculating Tokens TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def safe_eval(x):
    try:
        return ' '.join(eval(x)) if isinstance(x, str) else ''
    except:
        return ''

# Preparing the tokens for TF-IDF
# Combining all tokens into a single string for each tweet
df_result['tokens_combined'] = df_result['tokens'].apply(safe_eval)

# Initializing TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(df_result['tokens_combined'])

# Summarizing the TF-IDF scores for each tweet into a single relevance score
# Taking the mean TF-IDF score across all tokens for each tweet as a simple measure of content relevance
df_result['content_relevance'] = tfidf_matrix.mean(axis=1).A.flatten()

# Displaying the updated DataFrame with the Content Relevance scores
df_result[['tokens_combined', 'content_relevance']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalizing the sentiment scores to a 0-1 scale
scaler = MinMaxScaler()
df_result['normalized_sentiment'] = scaler.fit_transform(df_result[['sentiment_score']])

# Calculating the Engagement Score as an equal-weighted sum of likes, retweets, replies, and quotes
df_result['engagement_score'] = df_result[['like_count', 'retweet_count', 'reply_count', 'quote_count']].mean(axis=1)

# Estimating Network Influence
# Assuming each mention contributes equally and presence of URLs contributes to the score
df_result['mentions_count'] = df_result['mentions'].apply(lambda x: len(eval(x)) if pd.notnull(x) else 0)
df_result['urls_count'] = df_result['urls'].apply(lambda x: len(eval(x)) if pd.notnull(x) else 0)
df_result['network_influence'] = df_result['mentions_count'] + df_result['urls_count']

df_result['content_virality'] = df_result['retweet_count'] + df_result['quote_count']

# Adjusting the Social Capital Score calculation to include AI (using 'mentions_count') and CV ('content_virality')
# New weights for each component including AI and CV
new_weights = {'SI': 0.20, 'ES': 0.20, 'CR': 0.20, 'NI': 0.20, 'AI': 0.10, 'CV': 0.10}

df_result['social_capital_score_new'] = (
    new_weights['SI'] * df_result['normalized_sentiment'] +
    new_weights['ES'] * df_result['engagement_score'] +
    new_weights['CR'] * df_result['content_relevance'] +
    new_weights['NI'] * df_result['network_influence'] +
    new_weights['AI'] * df_result['mentions_count'] +
    new_weights['CV'] * df_result['content_virality']
)

# Displaying the updated DataFrame with the new Social Capital Score including graph metrics
df_result[['normalized_sentiment', 'engagement_score', 'content_relevance', 'network_influence', 'mentions_count', 'content_virality', 'social_capital_score_new']].head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Function to safely parse tokens and hashtags
def safe_eval(val):
    try:
        return eval(val)
    except:
        return []

# Combine tokens and hashtags into a single string for each tweet
def combine_tokens_hashtags(row):
    tokens = ' '.join(safe_eval(row['tokens']))
    hashtags = ' '.join(safe_eval(row['hashtags'])) if pd.notnull(row['hashtags']) else ''
    return tokens + ' ' + hashtags

# Apply the function to combine tokens and hashtags
df_result['combined_text'] = df_result.apply(combine_tokens_hashtags, axis=1)

# Creating TF-IDF matrix for combined text
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix_combined = tfidf_vectorizer.fit_transform(df_result['combined_text'])

# Convert TF-IDF matrix to DataFrame for easier manipulation
tfidf_df = pd.DataFrame(tfidf_matrix_combined.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Extract relevant metrics from the updated_df
metrics_df = df_result[['sentiment_score', 'diversity_score', 'context_score', 'like_count', 'reply_count', 'quote_count', 'retweet_count', 'impression_count']]

# Concatenate the metrics with the TF-IDF DataFrame
combined_matrix = pd.concat([metrics_df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# Normalize the combined matrix
scaler = StandardScaler()
normalized_matrix = scaler.fit_transform(combined_matrix)

# Proceed to apply PCA
pca = PCA(n_components=0.95)  # Retain 95% of the variance
pca_matrix = pca.fit_transform(normalized_matrix)

# Checking the shape of the PCA-transformed matrix and the variance explained
pca_matrix_shape = pca_matrix.shape
variance_explained = pca.explained_variance_ratio_.sum()

pca_matrix_shape, variance_explained
```

### Evaluation

```python
df_result['pca1_score'] = pca_matrix[:, 0]
df_result.head()
```

```python
recommendations_df = pd.read_csv('./Database/updated_recommendations.csv')

recommendations_with_pca1 = recommendations_df.merge(df_result[['item_id', 'pca1_score', 'author_id']], on='item_id', how='left')

recommendations_with_pca1.head()
```

```python
newRecommendations = pd.DataFrame()
for index, row in recommendations_with_pca1.iterrows():
    user_id = row['user_id']
    who_published = row['author_id']
    tweet_id = row['item_id']
    algorithm = row['algorithm']
        
    rating = recommendations_with_pca1[(recommendations_with_pca1['item_id'] == tweet_id) & (recommendations_with_pca1['algorithm'] == algorithm) & (recommendations_with_pca1['user_id'] == user_id)]['rating'].values[0]
 
    score = df_result[df_result['item_id'] == tweet_id]['pca1_score'].values[0]
    newRecommendations.at[index, 'item_id'] = tweet_id
    newRecommendations.at[index, 'author_id'] = who_published
    newRecommendations.at[index, 'user_id'] = user_id
    newRecommendations.at[index, 'ranking'] = 0
    newRecommendations.at[index, 'rating'] = rating
    newRecommendations.at[index, 'score'] = score
    newRecommendations.at[index, 'algorithm'] = f'{algorithm}-SCSA_PLUS_V3'
```

```python
ranking = 1
sortRecommendationsByUserAndAlgorithm = newRecommendations.sort_values(by=['user_id','algorithm', 'score'], ascending=False)

for index, row in sortRecommendationsByUserAndAlgorithm.iterrows():
    sortRecommendationsByUserAndAlgorithm.at[index, 'ranking'] = ranking
    if(ranking == 10):
         ranking = 1
    else: 
         ranking = ranking + 1

print(sortRecommendationsByUserAndAlgorithm.head())
sortRecommendationsByUserAndAlgorithm.drop('score', axis=1, inplace=True)

combined_df = pd.concat([recommendations_with_pca1, sortRecommendationsByUserAndAlgorithm], ignore_index=True)
combined_df.to_csv('./Database/output_v3.csv', index=False)
print()
```
